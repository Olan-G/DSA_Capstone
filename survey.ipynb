This program asks 10 survey questions and stores the results in an XLSX file <br>

```python
import pandas as pd
import os

# List of questions
questions = [
    "How often do you take short breaks (e.g., standing, stretching, walking) during your workday?",
    "Do you feel that your workload allows you to maintain a healthy work-life balance?",
    "How would you rate the availability of healthy food and drink options at the workplace?",
    "Do you engage in any physical activity during the workweek (e.g., walking, stretching, exercise)?",
    "How supportive do you feel your workplace is toward your physical and mental well-being?",
    "How often do you feel stressed or overwhelmed due to work responsibilities?",
    "Are you aware of any health and wellness programs offered by the company?",
    "Have you ever used any health-related resources provided by the company (e.g., counseling, gym access, wellness workshops)?",
    "Do you feel the physical workspace (e.g., lighting, ventilation, seating) supports your overall health?",
    "What suggestions do you have for improving health and wellness in the workplace?"
]

# List of answers
answers = [
    ['Never', 'Rarely', 'Sometimes', 'Often', 'Always'],
    ['Strongly disagree', 'Disagree', 'Neutral', 'Agree'],
    ['Very Poor', 'Poor', 'Fair', 'Good', 'Excellent'],
    ['No, never', 'Rarely (1 day/week)', 'Sometimes (2-3 days/week)', 'Often (4-5 days/week)', 'Daily'],
    ['Not at all supportive', 'Slightly supportive', 'Moderately supportive', 'Very supportive', 'Extremely supportive'],
    ['Never', 'Rarely', 'Sometimes', 'Often', 'Always'],
    ['Yes', 'No', 'Not sure'],
    ['Yes', 'No', 'I wasn\'t aware these existed'],
    ['Strongly disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree'],
]

# Check if the excel file already exists in the folder
if os.path.exists(os.path.join(os.getcwd(),"survey.xlsx")):
    # if yes, use the current file as the dataframe
    df = pd.read_excel('survey.xlsx')
else:
    # Else make the dataframe to store the questions and answers
    df = pd.DataFrame(columns = questions)

def save_responses(responses):
    # convert numerical to text
    for index, i in enumerate(responses):
        if index == 9: break
        else: responses[index] = answers[index][int(i)-1]
    df.loc[len(df)] = responses

responses = []

print("Workplace Health Survey Questioannaire")
print("Please enter the number of your answer for each question.")

while True:
    for i in range(len(questions)-1):
        print(f"{i+1}. {questions[i]}")
        answerLen = len(answers[i])
        for j in range(answerLen):
            print(f"   {j+1}.{answers[i][j]}")
        res = input()
        print(f"=========> Your Answer: {answers[i][int(res)-1]}")
        responses.append(res)
    # print last question
    print(f"10. {questions[9]}")
    res = input()
    responses.append(res)
    save_responses(responses)
    responses = []
    state = input("Answer another questionnaire? (yes/no)").lower()
    if state != 'yes':
        print("Survey response saved.")
        break

# Save the dataframe into an excel file
df.to_excel("survey.xlsx", index=False)
```
